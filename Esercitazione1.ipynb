**Esercizio 2**\
Valutare la funzione $f(x)=\frac{e^x-1}x$ in $x=10^{-n}$ per $n=1,2,...,16$.

```python
def espressione_con_cancellazione(x):
    return (np.exp(x)-1)/x
# espressione_con_cancellazione = lambda x: (np.exp(x)-1)/x ## alternativa


x = 10.**(-np.arange(1,17))
# x = np.asarray([10**(-n) for n in range(1, 17)]) ## alternativa 
y = espressione_con_cancellazione(x)
```

Successivamente riformulare le funzioni assegnate al fine di evitare il fenomeno della cancellazione numerica e, assumendo come valori esatti quelli che si ottengono mediante la riformulazione proposta, calcolare i corrispondenti errori relativi e confrontarli con la precisione di macchina.

```python
def espressione_esatta(x, n):
    sviluppo = 0
    # sommatoria da 1 a x^(n-1)/n!
    for i in range(1, n+1):
        sviluppo += x**(i-1)/np.math.factorial(i)
    return sviluppo

y_esatta = espressione_esatta(x, 15)
errore_relativo = np.abs(y-y_esatta)/np.abs(y_esatta)
print("Errore:",errore_relativo)
```

Per ogni valore di x, stampare e rappresentare graficamente il corrispondente errore relativo.

```python
data = np.stack([x, y, y_esatta, errore_relativo], axis=1)
print("Errore cancellazione numerica:\n", data)

plt.xscale('log')
plt.yscale('log')
plt.title("Errore cancellazione numerica")
plt.xlabel("x")
plt.ylabel("Errore")
plt.scatter(x, errore_relativo)
plt.show()
```

**Esercizio 4**

Data la funzione $f(x)=\sin(x)$, si approssimi $f'(x)$ con il seguente rapporto incrementale
$$
f'(x)\approx r(x)=\frac{\sin(x+h)-\sin(x)}{h}
$$
scegliendo $h=2^{-k}$ per $k=1,2,...,50$.

```python
def rapporto_incrementale(x, h):
    return (np.sin(x+h) - np.sin(x))/h

h = np.asarray([2**(-k) for k in range(1,51)])
x = np.pi

r = rapporto_incrementale(x, h)
print(np.stack([h, r], axis=1))
```

Fissare, per esempio, $x=\frac\pi4$ e rappresentare graficamente l'errore relativo
$$
\frac{|f'(x)-r(x)|}{|f'(x)|}
$$
e commentare i risultati

```python
x = np.pi/4
derivata_esatta = np.cos(x)

r = rapporto_incrementale(x, h)
errore_rapporto_incrementale = np.abs(derivata_esatta - r) / np.abs(derivata_esatta)

plt.xscale('log')
plt.yscale('log')
plt.title("Errore rapporto incrementale")
plt.xlabel("x")
plt.ylabel("Errore")
plt.scatter(h, errore_rapporto_incrementale)
plt.show()
```

Successivamente riformulare $r(x)$ al fine di evitare il fenomeno della cancellazione numerica e rappresentare graficamente il corrispondente errore relativo.

\sin(\alpha)-\sin(\beta) = 2 \sin \frac{\alpha-\beta}{2} \cos \frac{\alpha+\beta}{2}

$$
r(x) = \frac{\sin(x+h)-\sin(x)}{h} = 2 \sin \frac{x + h - x}{2} \cos \frac{x + h + x}{2} = 2 \sin \frac{h}{2} \cos (x + \frac{h}{2})
$$*
